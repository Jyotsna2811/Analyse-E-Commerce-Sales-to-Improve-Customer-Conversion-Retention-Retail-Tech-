```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
product_df=pd.read_csv('products_cleaned.csv')
order_df=pd.read_csv('orders_cleaned.csv')
order_item_df=pd.read_csv('order_items_cleaned.csv')
customer_df=pd.read_csv('customers_cleaned.csv')
geolocation_df=pd.read_csv('geolocation_cleaned.csv')
order_reviews_df=pd.read_csv('orders_reviews_cleaned.csv')
cancelled_orders_df=pd.read_csv("orders_cancelled.csv")
```

```python
order_item_df['discounted_price']=order_item_df['unit_price']*(order_item_df['discount(%)']/100)
```

```python
product_df['profit']=product_df['selling_price']-product_df['cost_price']
```

```python
product_df['Total_profit']=product_df['profit']-order_item_df['discounted_price']+order_item_df['shipping_cost']
```

```python
order_item_df['final_cost']=order_item_df['unit_price']-order_item_df['discounted_price']+order_item_df['shipping_cost']
```

```python
cancelled_orders_df.to_csv("cancelled_orders.csv",index=False)

customer_df.to_csv("Customers.csv", index=False)

order_df.to_csv("Orders.csv", index=False)

order_item_df.to_csv("Order_items.csv", index=False)

product_df.to_csv("Products.csv", index=False)

order_reviews_df.to_csv("order_reviews.csv", index=False)

geolocation_df.to_csv("Geolocation.csv", index=False)
```
